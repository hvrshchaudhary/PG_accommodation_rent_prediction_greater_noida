```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("./rawdataset.csv")
df.head()
```

```python
# get the number of missing data points per column
print(df.isnull().sum())
missing_val_count = df.isnull().sum()
```

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_val_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
# one reason behind null entries can be that the rooms did not have the following amenities so the owners didn't care
# to fill them (because if you're giving a service and you can charge extra for it, then why wouldn't you mention it)
# hence it makes sense to set them to false
df.fillna(False, inplace = True)
```

```python
print(df.isnull().sum())
```

```python
print(df.loc[[34]])
```

```python
# how many unique values do we have in this column?
print(df['Occupancy'].unique())
```

```python
# mapping each unique feature to an integer value
# df.Occupancy = df.Occupancy.astype("category").cat.codes
df.head()
```

```python
df['Occupancy'] = df['Occupancy'].replace({'DOUBLE' : 2, 'SINGLE' : 1, 'THREE' : 3, 'FOUR' : 4})
```

```python
# the numeric values 
print(df['Occupancy'].unique())
```
